```python
import sys
import pandas as pd
import yaml

# Cập nhật đường dẫn trước khi import các module nội bộ của dự án
sys.path.append('..')
from src.models.supervised import split_and_balance_data, train_and_evaluate_models

# 1. Đọc file config 
with open("../configs/params.yaml", "r", encoding="utf-8") as file:
    config = yaml.safe_load(file)

# 2. Đọc tập dữ liệu SẠCH 
processed_path = "../" + config['data']['processed_path']
df_clean = pd.read_csv(processed_path)

# 3. Chạy pipeline huấn luyện
X_train, X_test, y_train, y_test = split_and_balance_data(
    df_clean, 
    target_col='y', 
    test_size=config['experiment']['test_size'], 
    random_state=config['experiment']['random_state']
)

# 4. Huấn luyện và xem bảng xếp hạng mô hình
models, df_results = train_and_evaluate_models(X_train, X_test, y_train, y_test)
display(df_results)
```
